```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.model_selection import cross_val_score

df3 = pd.read_csv('df_imp.csv')
df4 = pd.read_csv('df_attr_select.csv')
```

```python
df3= df3.rename({'appCat.builtin': 'builtin'} , axis = 'columns')
df3= df3.rename({'appCat.communication': 'communication'} , axis = 'columns')
df3= df3.rename({'appCat.entertainment': 'entertainment'} , axis = 'columns')
df3= df3.rename({'appCat.other': 'other'} , axis = 'columns')
df3= df3.rename({'appCat.social': 'social'} , axis = 'columns')
df3= df3.rename({'appCat.unknown': 'unknown'} , axis = 'columns')
df3= df3.rename({'appCat.utilities': 'utilities'} , axis = 'columns')
df3= df3.rename({'appCat.finance': 'finance'} , axis = 'columns')
df3= df3.rename({'appCat.office': 'office'} , axis = 'columns')
df3= df3.rename({'appCat.travel': 'travel'} , axis = 'columns')
df3= df3.rename({'appCat.weather': 'weather'} , axis = 'columns')
df3= df3.rename({'appCat.game': 'game'} , axis = 'columns')
```

```python
df3.head()
```

```python
X = df3[['activity' , 'builtin' , 'communication' , 'entertainment' , 'finance' , 'game', 'office' , 'other' , 'social', 'travel' , 'unknown' , 'utilities', 'weather', 'call' , 'circumplex.arousal' , 'circumplex.valence' , 'mood' , 'screen' , 'sms']]
y = df3[['target']]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
y_train = y_train.values.ravel()
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics  
print()
# scores = cross_val_score(clf, X, y, cv=5)

# print("acc cv is:", (scores.mean(), scores.std()))
# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

# feature_imp = pd.Series(clf.feature_importances_,index=df3).sort_values(ascending=False)
# feature_imp

labels = ['activity' , 'builtin' , 'communication' , 'entertainment' , 'finance' , 'game', 'office' , 'other' , 'social', 'travel' , 'unknown' , 'utilities', 'weather', 'call' , 'circumplex.arousal' , 'circumplex.valence' , 'mood' , 'screen' , 'sms']


for feature in zip(labels, clf.feature_importances_):
    print(feature)

importances1 = clf.feature_importances_
sorted_indices1 = np.argsort(importances1)[::-1]

import matplotlib.pyplot as plt

 
plt.title('Feature Importance: All variables')
plt.bar(range(X_train.shape[1]), importances1[sorted_indices1], align='center')
plt.xticks(range(X_train.shape[1]), X_train.columns[sorted_indices1], rotation=90)
plt.tight_layout()
plt.show()
```
